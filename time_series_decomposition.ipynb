```python
import sys
import statsmodels as sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

% matplotlib inline
```

### Contruct trend

```python
time = np.arange(1, 51)
```

```python
trend = time * 2.75
```

```python
plt.plot(time, trend, 'b.')
plt.title("Trend vs Time")
plt.xlabel("minutes")
plt.ylabel("eletricity demand")
```

### Construct Seasonality

```python
seasonal = 10 + np.sin(time) * 10
```

```python
plt.plot(time, seasonal, 'g-.')
plt.title("Seasonality vs Time")
plt.xlabel("minutes")
plt.ylabel("eletricity demand")
```

### Construct Residual

```python
np.random.seed(10)
residual = np.random.normal(loc=0.0, scale=1, size=len(time))
```

```python
plt.plot(time, residual, 'r-.')
plt.title("Residuals vs Time")
plt.xlabel("minutes")
plt.ylabel("eletricity demand")
```

# Section 2: Aggregate Trend, Seasonality and Resisual Components

### Additive Time Series

Remember the equation for additive time series is simply O = T + S + R

```python
additive = trend +seasonal + residual
```

```python
plt.plot(time, additive, 'k.')
plt.title("Additive Time Series")
plt.xlabel('minutes')
plt.ylabel('eletricity deman')
```

### Multiplicative Time Series
Remember the equation for multiplicative time series is simply: O = T * S * R

```python
# ignoring residual to make more apparent
ignore_residual = np.ones_like(residual)
multiplicative = trend * seasonal * ignore_residual
```

```python
plt.plot(time, multiplicative, 'k-.')
plt.title("Multiplicative Time Series")
plt.xlabel('minutes')
plt.ylabel('eletricity deman')
```
